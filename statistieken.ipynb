### Wat is de maximale gemiddelde temperatuur in Rotterdam

Om dit te kunnen bereiken gaan we de max pandas functie gebruiken.


### Wat is de maximale (hoogst) gemiddelde temperatuur in Rotterdam?

De hoogste temperatur in Rotterdam is de doel van dit oefenen.

```python
is_Rotterdam = df["City"] == "Rotterdam"
avg_temp_Rotterdam = df["AverageTemperature"][is_Rotterdam].max()
print(avg_temp_Rotterdam)
```


### Wat is de minimale (laagste) gemiddelde temperatuur in Maastricht?

De laagste temperatur in Maastricht is de doel van dit oefenen.


```python
is_Maastricht = df["City"] == "Maastricht"
avg_temp_Maastricht = df["AverageTemperature"][is_Maastricht].min()
print(avg_temp_Maastricht)
```

### Hoeveel verschillende steden zitten in de dataset?

Hoevel unique steden zitten in deze dataset.

```python
count_unique_cities = df["City"].nunique()
print(count_unique_cities)
```

### Welke gemiddelde temperatuur komt het meeste voor? Rond hiervoor de temperatuur af op hele getallen.

Wat is de meest voorkomende temperatuur in deze dataset.


```python
# dataframe.mode()['Column'][0]
# round it up
most_frquent_avgTemperature = df["AverageTemperature"].round().mode().max()

print(most_frquent_avgTemperature)
```

### Wat zijn de top 3 warmste temperaturen gemeten? Let op: het eindresultaat moet bestaan uit 3 unieke temperaturen.

Waar kunnen wij de warmste temperaturen vinden, allen 3 top resultaten!


```python
# Net zoals verwacht de waarden zijn:
# 22.812
22.621
21.983
warmste_temp = df['AverageTemperature'].round(3).sort_values(ascending=False).head(4)

print(warmste_temp.drop_duplicates())
```
