### CSV bestaand lezen naar een dataframe

Laad het bestand in een Pandas dataframe. Zet het argument index_col op 'index'.

```python
import pandas as pd

# Load the dataset into a pandas dataframe
df = pd.read_csv('gdrive/My Drive/Colab Notebooks/bulldozers.csv')
```

### Histogram-outliers in Bulldozers

Maak een plot van de kolom 'YearMade' om te kijken of je outliers ziet. Je kunt hiervoor een histogram gebruiken. Zorg dat je een redelijk aantal bins kiest.

```python
# Plot a histogram of the 'YearMade' column
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['YearMade'], bins=30, edgecolor='black')
plt.title('Het aantal bulldozers dat in een periode gemaakt is')
plt.xlabel('Jaar')
plt.ylabel('Aantal bulldozers')
plt.show()
```

### Outliers weg!

Zijn alle waarden redelijke jaartallen voor het productiejaar van bulldozers? Er werden in het jaar 1000 nog géén bulldozers gemaakt. Dit zijn dus duidelijk outliers! Verwijder deze rijen uit de dataset en maak een nieuwe histogram van de overgebleven data.

```python
# Remove rows with 'YearMade' equal to 1000
df_cleaned = df[df['YearMade'] != 1000]

# Plot a histogram of the cleaned 'YearMade' column
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['YearMade'], bins=30, edgecolor='black')
plt.title('Het aantal bulldozers dat in een periode gemaakt is')
plt.xlabel('jaar')
plt.ylabel('Aantal bulldozers')
plt.show()

```
