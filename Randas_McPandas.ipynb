### Gezonde Maaltijd

Vind per maaltijdcategorie de maaltijd met het minst aantal calorieÃ«n.

```python
# Find the category with the minimum calories
min_calories_category = df['Calories'].idxmin()

# Print the category
print("Category with the least calories:", df.loc[min_calories_category, 'Category'])
```

### Barchart met het minst aantal calories per maaltijd, per categorie.

In de chart moet duidelijk zijn om welke categorie, en om welke maaltijd het gaat.



```python
# Find the rows with the minimum calories per category
min_calories_per_category = df.groupby('Category')['Calories'].idxmin()

# Create a new DataFrame containing the rows with the minimum calories
min_calories_df = df.loc[min_calories_per_category, ['Category', 'Item', 'Calories']]

# Create a vertical bar chart
bars = plt.bar(min_calories_df['Category'] + ':' + min_calories_df['Item'], min_calories_df['Calories'])

# Add labels and title (translated to Dutch)
plt.xlabel('Categorie - Maaltijd met Minste CalorieÃ«n')
plt.ylabel('Aantal calorieÃ«n')
plt.title('Aantal calorieÃ«n voor maaltijden met het minst aantal calorieÃ«n per categorie')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```
